```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_bicis = pd.read_csv('/home/bigdata/BicisCorunha/datos_citybikes/bicicorunha_data.csv', parse_dates=True)
```

```python
df_bicis
```

```python
df_bicis.info()
```

```python
df_bicis = df_bicis.drop(["Unnamed: 0"], axis=1)
```

```python
#Elimina rexistros duplicados.

df_bicis = df_bicis.drop_duplicates()
```

```python
df_bicis.info()
```

```python
df_bicis
```

```python
df_bicis["timestamp"] = df_bicis["timestamp"].str.rstrip('Z')

df_bicis["timestamp"] = pd.to_datetime(df_bicis['timestamp'], errors='coerce')

df_bicis
```

```python
df_tiempo = df_bicis[["timestamp"]]

df_tiempo.info()
```

```python
df_tiempo.drop_duplicates()
```

```python
df_tiempo["Diferencia"] = df_tiempo["timestamp"].diff()
```

```python
df_tiempo['tiempo_diferencia_segundos'] = df_tiempo["Diferencia"].dt.total_seconds()

df_tiempo
```
